```python
import pandas as pd
import numpy as np

def process_data(data, filtered_data, tickers):
    open_prices = data['Open'][tickers].values
    close_prices = data['Close'][tickers].values
    
    valid_mask = (open_prices != None) & (close_prices != None)
    returns = np.where(valid_mask, (close_prices - open_prices) / open_prices, None)
    
    for idx, tickers in enumerate(tickers):
        filtered_data[idx].extend(returns[:, idx][~np.isnan(returns[:, idx])])
    
    return filtered_data
```

```python
filtered_data = process_data(data, filtered_data, tickers)
```

```python
tickers = list(map(lambda x: x[1] ,data.columns))
tickers = tickers[1:]
open_prices = data['Open'][tickers].values
close_prices = data['Close'][tickers].values


valid_mask = (open_prices != None) & (close_prices != None) & (open_prices > 0) & (close_prices > 0)

returns = np.where(valid_mask, ((close_prices - open_prices) / open_prices) *100, 0)
```

```python
def map_floats(value):
    if value <= -10000:
        return 0
    elif value >= 10000:
        return 401
    else:
        return (value // 50)+201
```
